## Missing Values

```python
## Check the percentage of nan values present in each feature 
## 1-step make the list of features which has missing values
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]
```

```python
##2-step print the feature name and the percentage of missing values
for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(),4),'% missing values')
## these missing values are handled in feature engineering
```

### Since there are many missing values, we need to find the relationsip between missing value and Sales Price


```python
# make a copy
for feature in features_with_na:
    data = dataset.copy()
# make a variable that indicates the 1 if the observation was missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(),1,0)

#calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here with the relation between the missing values and the dependent variable is clearly visible that features having nan values are affecting the sales price. So we need to replace values with something meaningful which we will do in the Feature Engineering. 

Missing values has higher median than the non missing values

__Tempral Variable : (Ex. Datetime Variable)__

From the Dataset we have 4 year variables. We have to extract information from the datetime variables like no. of years or no. of days. One example in this specific scenerio can be difference in years between the house was built and the year the house was sold. We will be performing this analysis in the Feature Engineering.

```python
# List of variables that contain your information 
year_feature = [feature for feature in numerical_features if "Yr" in feature or "Year" in feature]
year_feature
```

```python
# Explore the content of these year variable
for feature in year_feature:
    print(feature, year_feature)
```

```python
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
# Analysis of the Temporal Variables
# Check whether there is any relation between year the house is sold and SalePrice i.e the output

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price v/s YearSold")
```

Here we can observe that the price is decreaing with respect to the years

```python
year_feature
```

```python
# Compare the difference between All years feature with SalePrice
for feature in year_feature:
    if feature!='YrSold':
        data = dataset.copy()
        # difference in year variable and year the house Sold
        data[feature] = data['YrSold'] - data[feature]
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
        
```
